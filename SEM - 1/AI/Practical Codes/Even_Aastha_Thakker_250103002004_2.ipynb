```python
import pandas as pd
import numpy as np

# DATASET MAKING
data = {
    "User_ID": ["DS01","DS02","DS03","DS04","DS05","DS06","DS07","DS08","DS09","DS10"],
    "Failed_Attempts": [3, 7, 2, 7, 1, 4, 6, 10, 8, 5],
    "Successful_Attempts": [12, 9, 15, 6, 18, 11, 10, 7, 14, 13]
}
df = pd.DataFrame(data)

# Save the file
df.to_csv("user_security_logs.csv", index=False)

print("CSV File Generated Successfully!")
print(df)
```

```python
# A) NumPy array for Failed_Attempts and mean and standard deviation
failed_attempts_array = np.array(df["Failed_Attempts"])
mean_failed_attempts = np.mean(failed_attempts_array)
std_failed_attempts = np.std(failed_attempts_array)

print(f"\nNumPy array for Failed Attempts: {failed_attempts_array}")
print(f"Mean of Failed Attempts: {mean_failed_attempts:.2f}")
print(f"Standard Deviation of Failed Attempts: {std_failed_attempts:.2f}")
```

```python
import matplotlib.pyplot as plt

df = pd.read_csv("user_security_logs.csv")

# C) Plotting graph
plt.figure(figsize=(8,5))
plt.bar(df["User_ID"], df["Failed_Attempts"])
plt.xlabel("User ID")
plt.ylabel("Failed Login Attempts")
plt.title("Failed Login Attempts per User")
plt.show()
```
