```python
# Reverse num using while loop

num = 123
print(num)

rev_num = 0
while number > 0:
    digit = number % 10  # Get the last digit
    rev_num = rev_num * 10 + digit  # Append the digit to the reversed number
    number = number // 10  # Remove the last digit from the original number
```
