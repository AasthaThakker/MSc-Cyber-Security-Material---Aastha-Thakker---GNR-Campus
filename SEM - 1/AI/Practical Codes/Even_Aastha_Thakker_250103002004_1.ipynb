```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# DATASET MAKING
data = {
    "Date": [
        "2025-02-01","2025-07-02","2025-06-03","2025-01-04","2025-04-05","2025-02-06","2025-08-07","2025-01-08","2025-03-09","2025-05-10"
    ],
    "Malware_Detected": [10, 14, 19, 7, 22, 25, 18, 30, 21, 16],
    "Files_Scanned": [480, 500, 520, 510, 540, 560, 550, 600, 620, 630]
}

df = pd.DataFrame(data)
df.to_csv("malware_logs.csv", index=False)
print("Dataset Created:\n", df)
```

```python
# A) NumPy array for malware detections
detections = np.array(df["Malware_Detected"])

mean_val = np.mean(detections)
std_val = np.std(detections)

print("\nMean Malware Detections:", mean_val)
print("Standard Deviation:", std_val)
```

```python
df2 = pd.read_csv("malware_logs.csv")

# B) First 5 rows
print("\nFirst 5 Rows:")
print(df2.head())

# B) Column Names
print("\nColumn Names:")
print(df2.columns)

# B) Summary Statistics
print("\nSummary Statistics:")
print(df2.describe())
```

```python
# C) Bar chart Malware vs Date
plt.figure(figsize=(8,5))
plt.bar(df2["Date"], df2["Malware_Detected"])
plt.xlabel("Date")
plt.ylabel("Malware Detected")
plt.title("Malware Detection per Day")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
